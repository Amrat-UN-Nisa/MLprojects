#Check the index of the duplicated value:

```python
# Check for duplicate values in each row
rows_with_duplicates = df[df.duplicated(keep=False)]

# Get the row indices of rows with duplicate values
indices_with_duplicates = rows_with_duplicates.index.tolist()

# Print the row indices of rows with duplicate values
print("Row Indices with Duplicate Values:")
print(indices_with_duplicates)
```

```python
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
df.rename(columns={'Performance Index':'performance in percentage' }, inplace=True)
```

```python
df.head()
```

#Label Encoding:

```python
#Create a LabelEncoder object
label_encoder = LabelEncoder()

# Encode the 'geography' column
df['Extracurricular Activities'] = label_encoder.fit_transform(df['Extracurricular Activities'])
```

```python
df.head(2)
```

```python
x = df.drop(['performance in percentage'],axis = 1)
y = df['performance in percentage']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state = 42)
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
predicted_train = lr.predict(x_train)
print("MSE:",mean_squared_error(y_train,predicted_train))
print("RMSE:",np.sqrt(mean_squared_error(y_train,predicted_train)))
print("r2_score:",r2_score(y_train, predicted_train))
```

```python
predicted_values = lr.predict(x_test)
```

```python
print("MSE:",mean_squared_error(y_test,predicted_values))
print("RMSE:",np.sqrt(mean_squared_error(y_test,predicted_values)))
print("r2_score:",r2_score(y_test, predicted_values))
```

```python
print("Training score:",lr.score(x_train,y_train))
```

```python
print("Testing score:",lr.score(x_test,y_test))
```
